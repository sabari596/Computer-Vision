```python
# Import libraries
import cv2 as cv #OpenCV library (for camera, video, image processing)

# Open the default camera (index 0).
# If you have multiple cameras, 1 = external USB cam, 2 = another device, etc.
cam = cv.VideoCapture(0)

while True:
    # Read one frame from the camera
    # if the frame was captured, then value of success will be True
    # the actual image will be in the variable frame
    success, frame = cam.read()

    # If no frame is captured, show error and continue loop
    if not success:
        print("Camera not detected")
        continue

    # Flip the frame horizontally (mirror effect)
    frame = cv.flip(frame, 1)

    # Display the frame in a window called "Camera preview"
    cv.imshow("Camera preview", frame)

    # Wait 1 millisecond for a key press. If the user presses 'q', exit the loop    
    if cv.waitKey(1) == ord('q'):
        break

# Release the camera (free the hardware)
cam.release()

# Close all OpenCV windows
cv.destroyAllWindows()
```

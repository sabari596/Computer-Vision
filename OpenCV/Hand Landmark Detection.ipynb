```python
import cv2 as cv
import mediapipe as mp
```

```python
# Initialize MediaPipe Hands
mpHands = mp.solutions.hands
drawing = mp.solutions.drawing_utils
```

```python
# Open the default camera
cam = cv.VideoCapture(0)

# Configure the MediaPipe Hands model
hands = mpHands.Hands(
    static_image_mode=False,          # Detect hands in live video
    max_num_hands=2,                  # Detect up to 2 hands
    min_detection_confidence=0.5,     # Minimum confidence for detection
    min_tracking_confidence=0.5       # Minimum confidence for tracking
)
```

```python
while True:
    # Read a frame from the camera
    success, frame = cam.read()
    if not success:
        print("Camera not detected")
        continue

    # Flip for mirror-like view
    frame = cv.flip(frame, 1)

    # Convert BGR  RGB because MediaPipe uses RGB
    frameRGB = cv.cvtColor(frame, cv.COLOR_BGR2RGB)

    # Process the frame to detect hands
    handsDetected = hands.process(frameRGB)

    # If hands detected
    if handsDetected.multi_hand_landmarks:
        for hand_landmarks in handsDetected.multi_hand_landmarks:
            # Print all detected hand landmarks
            print(hand_landmarks)

            # Draw hand landmarks and connections on the frame
            drawing.draw_landmarks(
                frame,
                hand_landmarks,
                mpHands.HAND_CONNECTIONS
            )

    # Show the frame with drawn landmarks
    cv.imshow("Hand Landmarks", frame)

    # Exit if user presses 'q'
    if cv.waitKey(1) == ord('q'):
        break

# Release camera and close windows
cam.release()
cv.destroyAllWindows()
```
